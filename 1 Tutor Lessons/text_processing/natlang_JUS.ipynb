```python
import os
import string
import glob
import pprint
import re

import os
print(os.getcwd())
from os.path import basename
```

```python
directory_path = r"C:\0 Python Docs\docs to load into python or nltk - JUS"

text_files = glob.glob(f"{directory_path}\\*.txt")
print(f"Number of .txt files: {len(text_files)}") 

for directory_path in text_files:
    pprint.pprint(os.path.basename(directory_path)) 

for file in text_files:
    pprint.pprint(file)


print()
print("########################################## This is working ##########################################")
print()
```

```python
# append all into one string variable, corpus1
import pprint
print()
print("This appends content of each file to corpus")
print()
corpus = "  "

for directory_path in text_files:                     
    with open(directory_path, 'r', encoding='utf-8', errors='ignore') as file:
        content = file.read()
        corpus += content                       # + '\n'            # adds new line between files // this wouldn't work this time, not sure why
        print()
        print(f"---- {os.path.basename(directory_path)} ---")           
        print(content[:200])                                                          #prints first 200 characters as a preview
print()
print(corpus)
print()
print("#######################################This is working#############################################")
print()
```

```python
# removing invalid characters and makes the text all lowercase

print("This checks for invalid characters and removes them.")
print()

valid_characters = set(string.ascii_lowercase + " -'@")

corpus5 = " "

for char in corpus4.lower():
    if char in valid_characters:
        corpus5 += char
    else:
        continue

pprint.pprint(corpus5)

print("############################################ This is working  ############################################")
```

```python
# word frequency

print("This is the word frequency calculation.")
print()

words = corpus5.split(" ")
frequency = {}
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

pprint.pprint(frequency)                                       # bc I grabbed most content from websites, I have plenty of non-words

print("################################# This is working #####################################")
```
