```python
# again - this time glob
import os
print(os.getcwd())
from os.path import basename

import pprint
import string
import glob

directory_path = r"C:\0 Python Docs\docs to load into python or nltk - IM"

text_files = glob.glob(f"{directory_path}\\*.txt")
print(f"Number of .txt files: {len(text_files)}")                     # count the number of files in a directory.  and add **\\ and set recursivev = True if there are subfolders you want to include

for directory_path in text_files:
    print("Full path: ", directory_path)                             # incl path and filename
    print("Basename: ", os.path.basename(directory_path))            # incl filename only


for file in text_files:
    print(file)

print("This is the last filename:")
print(os.path.basename(directory_path))               # lists the last folder in the directory_path, not the txt files.
 

print("This is working")

```

```python
print()
print("This appends content of each file to corpus")
print()
corpus = "beginning of string"

for directory_path in text_files:                     
    with open(directory_path, 'r', encoding='utf-8', errors='ignore') as file:
        content = file.read()
        corpus += content + '\n'            # adds new line between files
        print()
        print(f"---- {os.path.basename(directory_path)} ---")           
        print(content[:200])                                                          #prints first 200 characters as a preview
print()
print("beginning of corpus: ", corpus[:400])
print()
print("#######################################This is working#############################################")
print()


```

```python
print("This does line.strip")
print()

import string

cleaned_corpus = corpus.strip()

print(cleaned_corpus)
print()
print("######################################this is in process #############################")
print()
```

```python
print("This checks for invalid characters and removes them.")
print()

valid_characters = set(string.ascii_lowercase + " -'@")

cleaned_corpus2 = "beginning of string"

for char in cleaned_corpus.lower():
    if char in valid_characters:
        cleaned_corpus2 += char
    else:
        continue

pprint.pprint(cleaned_corpus2)

print("############################################ This is working  ############################################")
```

```python

print("This is the word frequency calculation.")
print()

words = cleaned_corpus2.split(" ")
frequency = {}
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

pprint.pprint(frequency)                                       # bc I grabbed most content from websites, I have plenty of non-words

print("################################# This is working #####################################")
```
