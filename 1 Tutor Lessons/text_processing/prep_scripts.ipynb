```python
import pprint
import string

valid_characters = string.ascii_lowercase + " -'@"

status = "parts of this are definitely working. need to use the text editor to review the output - otherwise crops"
print(status)

all_lines = []

with open('scripts.txt', 'r', encoding='latin-1') as file:   # tried encoding UTF-8 but got an error
    for line in file:
        cleaned_line = line.strip()                     #strip leading/trailing whitespace and process the line
        print(cleaned_line)
        all_lines.append(cleaned_line)

print("after this prints the fulltext of all the scripts - things are cropping so I cannot see")
fulltext = '\n'.join(all_lines)
#print(fulltext)

new_fulltext = ""
for letter in fulltext:
    if letter not in valid_characters:
        continue
    new_fulltext += letter
   
```

```python
# frequency & lowercase
print("frequency & lowercase working - but bc scripts, dialogue contains many non-complete words")
new_fulltext = new_fulltext.lower()

words = new_fulltext.split(" ")
frequency = {}
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

pprint.pprint(frequency)
```
